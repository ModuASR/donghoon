```python
device = 'cpu'

import os
import math
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
import torchaudio as ta, torchaudio.functional as taF

import matplotlib.pyplot as plt
import IPython.display as ipd
import yaml

import librosa

import time

from config import Config, ConfigList, ConfigDict, configurable
from spec import WavEmbed, Reverberation, STFT, ColoredNoise, STFT2Spec, Decibel, SpecAugment
```

```python
path_wav = 'KsponSpeech_000001.wav'
wav, sr = librosa.core.load(path_wav, sr=16000, mono=True)
wav = torch.tensor(wav).reshape(1, 1, -1)
wav.shape
```

```python
wav_embed = WavEmbed(add_noise=False, add_reverberation=False, add_spec_augment=False).to(device)
```

```python
spec, spec_length = wav_embed(wav.to(device))
print(spec.shape, spec_length)
plt.imshow(spec[0].cpu().numpy())
```

```python
wav_embed = WavEmbed(add_noise=True, add_reverberation=False, add_spec_augment=False).to(device)
```

```python
spec, spec_length = wav_embed(wav.to(device))
print(spec.shape, spec_length)
plt.imshow(spec[0].cpu().numpy())
```
